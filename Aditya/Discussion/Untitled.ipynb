```python
for con in failed_country:
    try:
        os.makedirs(con)
    except:
        pass
    for y in range(len(Years)):
        Country_matrix1 = pd.read_csv('/home/adityachourasiya/Desktop/Working/SGSIS/FNA/DataCountry' + Years[y] + '.csv')
        del Country_matrix1['ROW']
        Country_matrix1 = Country_matrix1.set_index('Unnamed: 0')
        Country_matrix1 = Country_matrix1.drop(index = 'ROW')
        Countries = list(Country_matrix1)
        node_size = np.diag(Country_matrix1)
        export = list(Country_matrix1[con])
        imprt = list(Country_matrix1.loc[con])
        trade_ratio = np.array(export)/np.array(imprt)
        idx = Countries.index(con)
        source = [con] * len(Countries)
        gdp_ratio = node_size/node_size[idx]
        data = pd.DataFrame()
        data['Source'] = source
        data['Country'] = Countries
        data['Trade Ratio'] = trade_ratio
        data['GDP Ratio'] = gdp_ratio
        data.to_csv(con + '/gdp_trade' + Years[y] + '.csv', index = False)
        
        
        imprt.remove(imprt[idx])
        export.remove(export[idx])
        s1 = list([idx] * (len(Countries)-1))
        s2 = list(range(len(Countries)))
        s2.remove(idx)
        src = s1 + s2
        tar = s2 + s1
        exp = export + imprt
        netw = pd.DataFrame()
        netw['Source'] = src
        netw['Target'] = tar
        netw['Weight'] = exp
        netw.to_csv(con + '/network' + Years[y] + '.csv', index = False)
        
        n = len(Countries)
        df = pd.DataFrame()
        df['Id'] = range(len(Countries))
        df['Label'] = list(Countries)
        df['Weight'] = node_size
        df.to_csv(con + "/nodes_" + Years[y] + ".csv", index=False)
```
